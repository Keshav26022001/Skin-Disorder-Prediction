# Data Preprocessing and Feature Engineering

## 1) PCA

```python
from sklearn.decomposition import PCA
# Separating the target variable and independent variables
X = data.iloc[:, :-1]
y = data.iloc[:, -1]

# Dropping the 'Age' Column
Age = data['Age']
X_no_Age = X.drop('Age', axis=1)
```

```python
# Applying PCA on the features excluding 'Age'
pca = PCA()
X_pca = pd.DataFrame(pca.fit_transform(X_no_Age))

# Generating dynamic column names for principal components
num_components = X_pca.shape[1]
X_pca.columns = [f'PC{i+1}' for i in range(num_components)]
```

```python
# Calculate explained variance and cumulative variance
explained_var = pca.explained_variance_ratio_
cum_var = np.cumsum(explained_var)
print(cum_var)
```

```python
# Selecting principal components that explain up to 90% variance
n_components_90 = np.argmax(cum_var >= 0.90) + 1
print(f"Number of components to retain 90% variance: {n_components_90}")
```

```python
X_pca_90 = X_pca.iloc[:, :n_components_90]
# Adding 'Age' column back to the PCA-transformed data
X_final = pd.concat([X_pca_90, Age.reset_index(drop=True)], axis=1)
```

```python
X_final.head()
```

```python
# Splitting the training and testing data
from sklearn.model_selection import train_test_split
X= X_final
y= y=data.iloc[:,-1]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

# Model Implementation

## 1) Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lreg_model = LogisticRegression(random_state = 42)
lreg_model.fit(X_train, y_train)
```

#### Here the model is not converged for the default number of iterations, so before evaluating the model we must find the best parametrs for the model.

```python
from sklearn.model_selection import GridSearchCV
# Define parameter grid
param_grid = {
    'solver': ['lbfgs', 'saga', 'newton-cg', 'liblinear'],
    'C': [0.001, 0.01, 0.1, 1, 10],
    'max_iter': [500, 1000, 2000],  # Increased iterations to improve convergence
    'tol': [1e-4, 1e-6],  # Lower tolerance for better optimization
    'class_weight': [None, 'balanced']  # Handles imbalanced data
}

# Initialize model
log_reg = LogisticRegression(random_state=42)

# GridSearchCV
grid_search = GridSearchCV(log_reg, param_grid, cv=5, scoring='accuracy', verbose=1, n_jobs=-1)
grid_search.fit(X_train, y_train)

# Print best parameters
print("Best Hyperparameters:", grid_search.best_params_)
```

```python
best_model = LogisticRegression(C=1, class_weight=None, max_iter=500, solver='liblinear', tol=0.0001, random_state=42)
best_model.fit(X_train, y_train)
```

```python
y_pred_lreg = best_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
print(classification_report(y_test, y_pred_lreg))
cm_lreg = confusion_matrix(y_test, y_pred_lreg)
sns.heatmap(cm_lreg, cmap = 'Reds', annot = True)
Accu_Sc_Log_reg = accuracy_score(y_test, y_pred_lreg)
print(f'Accuracy Score for the Logistic Regression is:\n{Accu_Sc_Log_reg}')
```

#### Perfect classification for classes 1, 5, and 6, with precision, recall, and F1-score = 1.00.

#### Class 4 has slightly lower precision & recall (87% & 93%), meaning occasional misclassification.

#### Macro & weighted averages (~97%) confirm consistent performance across all classes.

#### The model shows strong generalization with high recall, ensuring most classes are correctly identified.

```python
from sklearn.model_selection import GridSearchCV
para_SVM = { 'C' : [1, 10, 100], 'kernel': ['linear', 'poly', 'rbf'],
           'gamma' : ['scale', 'auto', 0.001, 0.01, 0.1, 1],  'max_iter' : [1000, 1500, 2000, 2500]}
SVM = SVC(random_state = 42)
grid_search = GridSearchCV( estimator = SVM, param_grid = para_SVM, cv = 5, n_jobs = -1, verbose =1)
grid_search.fit(X_train, y_train)
```

```python
best_SVM_model = grid_search.best_estimator_
best_SVM_model
```

```python
y_pred_best_svm = best_SVM_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
print(classification_report(y_test, y_pred_best_svm))
cm_svm = confusion_matrix(y_test, y_pred_best_svm)
sns.heatmap(cm_svm, cmap = 'Purples', annot = True)
Accu_Sc_svm = accuracy_score(y_test, y_pred_best_svm)
print(f'Accuracy Score for the SVM is:\n{Accu_Sc_svm}')
```
